# Data Visualization: Fatality Rates

## Setting up the workspace

The best way to understand our dataset is to visualize it. Seaborn and Matplotlib are going to be incredibly helpful for this goal. Landmark Airlines needs three recommendations of aircraft to invest in based on how safe the planes are. First, we must define what constitutes a "safe" aircraft.

Some ways we can determine the safety of aircraft are:

- Calculating the total fatality rate
- Analyzing fatality rates based on makes
- Examining fatality rates based on models

As Data Scientists, it is imperative that we represent this data in a clear way for businesses to understand and make informed decisions. We believe that the above visualizations will provide us with the clearest picture.

To begin, we must import the necessary libraries and CSV file(s).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# Ignores warnings 
warnings.filterwarnings("ignore")
```

```python
# Opening CSV file and assigning it to a Dataframe varrible
df = pd.read_csv('../data/cleaned_aviation.csv')
df.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
# Reading the DataFrame
df
```

Okay, now that we've managed to import and open the CSV file, we can begin creating visualizations based on the data we have.

Before diving into that, let's gather some information and understanding from the CSV file. This will help us identify the columns we want to utilize for our visualizations.

```python
# Gathing info on the DataFrame
df.info()
```

```python
# Lets make sure that we can claerly get affirmative data rather then any "Unavailable" data
injury_data = df[df['Injury.Severity'] != 'Unavailable']
```

## Creating the visualizations

To comprehend what is considered unsafe, we must initially grasp what qualifies as safe. With over 20,000 entries of data and minimal missing values, we have the capability to confidently calculate the average fatality rate for accidents within the United States. This calculation will provide us with a comprehensive understanding of which airplanes are more prone to higher fatalities and which ones tend to be safer and more sustainable.

```python
# Calculate average fatality rate
average_fatal = len(injury_data[injury_data['Injury.Severity'] == 'Fatal'].index) / len(injury_data.index)
print("Average fatality rate: " + str(round(average_fatal * 100, 2)) + '%')
```

Okay, so the average fatality rate in accidents within the USA is 18.23%. For the purpose of recommending makes and models to Landmark Airlines, anything with a fatality rate above 22% will be automatically excluded from consideration.

Now that we have the average fatality rate, our next step is to calculate the average fatality rate per make. To achieve this, we will create a bar graph illustrating the top 25 most frequently crashed makes.

### Fatality rate by make

```python
# Importing percent formatter
from matplotlib.ticker import PercentFormatter

# Extract list of 25 makes most frequently involved in accidents
make_top25 = df['Make'].value_counts().nlargest(25).index.tolist()

# Only keep data with a make in the top list
make_data = injury_data[injury_data['Make'].isin(make_top25)]

# Percentage fatality by make
makefatal = make_data.groupby('Make')['Injury.Severity'].value_counts(normalize = True) \
.unstack('Injury.Severity').reindex(make_top25)

makefatal.plot.bar(stacked = True, figsize=(18,5), width = 0.8, color = ['#003366','#2990EA'])

# Line with average fatality
plt.axhline(y = average_fatal, color = 'orange', linestyle = '-')
plt.axhline(y = .22, color = 'r', linestyle = '-')

#Formatting
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.title('Fatality rate by make')
plt.legend(title = "Injury Severity", loc='upper right')
plt.xlabel('Make of airplane')
plt.ylabel('Fatality Rate')

#Saving plot as .png file
plt.savefig("make_plot.png")
plt.show()
```

### Fatality rate by model

So what now? Do we simply conclude and advise Landmark Airlines to acquire Cessna and Piper aircraft and proceed with optimism?

Well, although fatality rates by make provide an initial measure of safety, it's crucial to emphasize that **this alone is insufficient**. Our stakeholders are seeking precise recommendations on which planes to invest in. They rely on our expertise and discernment to fulfill this task by offering three well-considered suggestions. Let's refine our focus and delve deeper into the intricacies of the Cessna and Piper models.

Our next step should involve examining the specific fatality rates within both of these makes.

```python
# Fatality rate of Cessna and Piper
cessna_injury_rates = df.loc[df['Make'] == 'Cessna']['Injury.Severity'].value_counts(normalize = True)
piper_injury_rates = df.loc[df['Make'] == 'Piper']['Injury.Severity'].value_counts(normalize = True)
cessna_fatality = cessna_injury_rates.values[1]
piper_fatality = piper_injury_rates.values[1]

print(f'Fatality rate of Cessna: {str(round(cessna_fatality * 100, 2))}%')
print(f'Fatality rate of Piper: {str(round(piper_fatality * 100, 2))}%')
print()
print("Average fatality rate: " + str(round(average_fatal * 100, 2)) + '%')
```

As indicated in the earlier bar chart, Cessna exhibits a significantly lower fatality rate compared to both Piper and the average fatality rate, which is certainly positive. Our progression is aligned with the correct trajectory.

However, Piper showcases a relatively higher fatality rate than Cessna, albeit not substantially surpassing the average. This disparity could be attributed to concealed variables that influence or skew the fatality rate. For instance, it's plausible that Piper has a generally low fatality rate, but the presence of one highly dangerous aircraft could distort the data's representation. It's possible that, overall, most Piper models are safer than most Cessna models.

This illustrates the necessity of delving deeper into the data before formulating our recommendations.

```python
cessna_models = df.loc[df['Make'] == 'Cessna']['Model'].value_counts().head(25)
piper_models = df.loc[df['Make'] == 'Piper']['Model'].value_counts().head(25)
```

```python
# Extract list of 25 Cessna models most frequently involved in accidents
ces_model_top25 = cessna_models.index.tolist()

# Only keep data with a model in the top list
ces_model_data = injury_data[injury_data['Model'].isin(ces_model_top25)]

# Percentage fatality by Cessna model
ces_model_fatal = ces_model_data.groupby('Model')['Injury.Severity'].value_counts(normalize = True) \
.unstack('Injury.Severity').reindex(ces_model_top25)

ces_model_fatal.plot.bar(stacked = True, figsize=(18,5), width = 0.8, color = ['#003366','#2990EA'])

# Line with average fatality
plt.axhline(y = cessna_fatality, color = 'orange', linestyle = '-')
plt.axhline(y = average_fatal, color = 'r', linestyle = '-')

#Formatting
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.title('Fatality rate by Cessna Model')
plt.legend(title = "Injury Severity", loc='upper right')
plt.xlabel('Model of airplane')
plt.ylabel('Fatality Rate')

#Saving plot as .png file
plt.savefig("cessna_model_plot.png")
plt.show()

```

# Data Visualization: Weather Conditions

It would be naive to assume that investing solely in aircraft with the lowest fatality rates would suffice. Numerous factors contribute to the survival rates of individuals within an aircraft, such as time of day, weather conditions, and pilot decisions. These variables significantly impact the statistics. However, we need a consistent and reliable parameter to base our decisions upon.

Among the array of factors, weather conditions emerge as a major concern. Flight delays during adverse weather conditions are well-founded, as these conditions can pose challenges during takeoff and landing, including the risk of severe turbulence. Within our dataset, two types of weather conditions are measured for our comprehension:

- IMC (Instrument Meteorological Conditions)
- VMC (Visual Meteorological Conditions)

VMC can be understood as "meteorological conditions expressed in terms of visibility," whereas IMC can be described as "weather conditions that necessitate pilots to primarily rely on flight instruments." It's reasonable to assume that flying in IMC conditions is riskier than in VMC conditions. However, we should examine the data to confirm this assumption.

By visualizing:

- The fatality rate of VMC and IMC flights compared to the average flight fatality rate
- The comparison between the volume of VMC and IMC flights in the dataset
- Aircraft damage based on weather conditions

We can hope to fully grasp the safest and most desirable flight conditions.

### Amount of VMC and IMC flights

Recognizing the scope of the data at hand is crucial. Our dataset centers on accidents within the United States. While a substantial volume of data on one side might provide insights into safety, it could equally reflect the prevalence of certain flight types. By initially comprehending the extent of the data, we can subsequently probe into the factors contributing to its distribution and clustering.

Lets create a cat plot to visualize what were working with.

```python
# Removing any unknown values
weather_data = df = df[df['Weather.Condition'] != 'Unknown']
```

```python
# Creating a cat plot
sns.catplot(y ="Weather.Condition",data = weather_data , kind= "count", palette = 'mako');

#Saving plot as .png file
plt.savefig("weather_count.png")
```

Wow thats alot of VMC flights, however whats the exact ratio of flight?

```python
# Checking the percentage of VMC/IMC flights
weather_data['Weather.Condition'].value_counts(normalize = True)
```

93% of accidents within the USA involve VMC flights. This observation raises two potential interpretations. First, VMC flights might be inherently risky, constituting the majority of accidents. Alternatively, this high percentage could indicate that most flights occur during VMC weather conditions, implying a preference for VMC.

Given our understanding of the data, we can deduce that VMC conditions are actually safer for flying. While the majority of flights do take place during VMC conditions, it's important to note that VMC implies clear visibility without reliance on aircraft instruments. This means no fog, no severe storms—nothing obstructing vision that would necessitate instrument use.

To confirm our inference, we should delve into the fatality rates associated with each condition. This will provide us with a more conclusive assessment of safety.

### Fatality rate of VMC and IMC flights

A simple bar graph with the average fatality rate of VMC and IMC flights compared to the average fatality rate of all flights within the nation should be perfect to visualize what we need.

```python
# Remove data without weather condition and injury severity
weather_data = injury_data[injury_data['Weather.Condition'] != 'Unknown']


# Stacked relative frequencies of accidents happening during certain weather conditions
weather_data.groupby('Weather.Condition')['Injury.Severity'].value_counts(normalize = True) \
.unstack('Injury.Severity').plot.bar(stacked = True, color = ['#003366','#2990EA'])

# Line with average fatality
plt.axhline(y = average_fatal, color = 'r', linestyle = '-')

#Formatting
plt.xticks(rotation = 0)
plt.title('Fatality rate during IMC or VMC')
plt.xlabel('')
plt.legend(title = "Injury Severity", loc='upper right')
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))

#Saving plot as .png file
plt.savefig("weather_plot.png")
```

It looks like we were correct; VMC had the largest amount of data, but it still had a lower fatality rate compared to IMC flights. Contextually, this makes sense; pilots are going to have a much easier time flying in visually clear conditions. During IMC flights, the pilot not only has to worry about not being able to see as clearly, but whatever is obstructing his vision can physically affect the aircraft as well. During thunder storms, they wouldn't be able to see much, but the rain and thunder, as well as potential humidity or other conditions, would affect how the actual flight of the plane goes as well.

# Data Visualization : Date time visuals

Understanding the temporal distribution of aviation accidents is essential for effectively managing risk and ensuring the safety of our operations. The interplay between different time periods, such as months and days of the week, can offer valuable insights into patterns of accidents, allowing us to make informed decisions and mitigate potential hazards.

Weather conditions exhibit seasonal variations throughout the year, which can significantly impact flight operations. Depending on the time of year, airports can be more or less busy. Major events in the year such as New Year's, Christmas, and other nationally celebrated holidays all contribute to the frequency of these crashes. By identifying the months with the highest accident occurrences, we can implement targeted strategies to enhance safety protocols during these periods. Additionally, recognizing the variability in flight volumes across different months empowers us to allocate resources more effectively and anticipate potential challenges.


We can do this by visualizing:
- Number of accidents per month since 1995
- Number of accidents per day of the week since 1995

```python
# Months with the most accidents
plot = sns.countplot(x = 'Month.Abbr', color = 'b', data = df)
plot.set(xlabel = 'Month', ylabel = None, title = 'Number of aviation accidents per month');

#Saving plot as .png file
plt.savefig("month_count.png")
```

There tends to be an influx of accidents that occur during the summer months. This is likely because most flights happen during this time, as this is when most people want to vacation with their friends and family, but we can't tell for sure without the use of another data set.

```python
# Days with the most accidents
plot = sns.countplot(x = 'Day.Name.Abbr', order = ['Mon','Tue','Wed','Thu','Fri','Sat', 'Sun'], color = 'b', data = df)
plot.set(xlabel = 'day', ylabel = None, title = 'Number of aviation accidents per day of the week');

#Saving plot as .png file
plt.savefig("day_count.png")
```

These visualizations provide insights into the distribution of accidents over months and days of the week, helping Landmark Airlines make informed decisions about scheduling flights and taking necessary precautions during certain periods. It appears that most accidents do occur during the weekends, but it's hard to see if there's a correlation; it is likely without causation. Most people will schedule their flights on the weekends so that they can leave or return to their day jobs and responsibilities.

We don't believe this data represents that weekends and summer months should be avoided; there simply isn't enough data for me to comfortably come to that conclusion. However,  this still tells us we should be aware of the frequencies of most flights. Being able to understand when the airspace will likely be most crowded will help us understand how many flights we can actually sell during a given month or day.
